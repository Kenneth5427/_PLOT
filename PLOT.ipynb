```python
import matplotlib.pyplot as plt
import numpy as np
import h5py
```

```python
f= h5py.File("kappa-m111111.hdf5")
list(f)
temp = f['temperature'][:]
kappa_0 = f['kappa'][:,0]
kappa_1 = f['kappa'][:,1]
kappa_2 = f['kappa'][:,2]
k = (kappa_0+kappa_1+kappa_2)/3
kappa_300 = k[30]
kappa_600 = k[60]
kappa_800 = k[80]
print("k_300 =",kappa_300,"\nk_600 =",kappa_600,"\nk_800 =",kappa_800)
```

```python
temp = f['temperature'][:]
kappa_0 = f['kappa'][:,0]
kappa_1 = f['kappa'][:,1]
kappa_2 = f['kappa'][:,2]
k = (kappa_0+kappa_1+kappa_2)/3
p1 = plt.plot(temp, kappa_0,linewidth=3.5 , label='kappa_xx')
p2 = plt.plot(temp, kappa_1, label='kappa_yy')
p3 = plt.plot(temp, kappa_2, label='kappa_zz')
p4 = plt.plot(temp, k, label='k')
l1 = plt.legend(loc='upper right')
plt.gca().add_artist(l1)
plt.xlim(300,1000)
plt.ylim(0,4)
plt.xlabel("temperature (K)")
plt.ylabel("kappa (W/K m)")
plt.title("lattice thermal conductivity")
plt.grid()
plt.savefig("l-thermalcond.jpg", dpi=600)
```

```python
# 300 K file reading list
file = open('300K.dat')
e_300 = []
s_300 = []
c_300 = []
k_300 = []
for line in file:
    e_300 += [line.split()]
    s_300 += [line.split()]
    c_300 += [line.split()]
    k_300 += [line.split()]

# file writing list (T = 300 K)
energy_300 = [e[0] for e in e_300]
seebeck_300 = [s[4] for s in s_300]
e_conduc_300 = [c[5] for c in c_300]
e_kappa_300 = [k[7] for k in k_300]

# we seperated the calculation process of power factor
pow_1_300 = [float(s)*float(s) for s in seebeck_300]
pow_2_300 = [float(q)*float(c) for q,c in zip(pow_1_300, e_conduc_300)]

# ZT for only eletron thermal conductivity
e_ZT_300 = [300*float(p)/float(k) for p,k in zip(pow_2_300, e_kappa_300)]

# ZT for elctron and lattice thermal conductivity
ZT_300 = [(300*float(p)*10**(-14))/(float(k)*10**(-14)+ kappa_300) for p,k in zip(pow_2_300, e_kappa_300)]

# transform the unit or form (T = 300 K)
E_300 = [float(i)*13.602 for i in energy_300]
S_300 = [float(i)*10**(6) for i in seebeck_300]
C_300 = [float(i)*10**(-19) for i in e_conduc_300]
P_300 = [float(i)*10**(-11) for i in pow_2_300]
```

```python
# plot electron conductivity
plt.plot(E_300, C_300, label='300 K')
plt.plot(E_600, C_600, label='600 K')
plt.plot(E_800, C_800, label='800 K')
l1 = plt.legend(loc='upper right')
plt.gca().add_artist(l1)
plt.xlim(-1, 1)
plt.ylim(0, 25)
plt.xlabel("Energy (eV)")
plt.ylabel("electron conductivity (W/K m)")
plt.title("Electron Conductivity")
plt.grid()
plt.savefig("elecond.jpg", dpi=600)
```
